```python
import pandas as pd
import geopandas as gpd
import numpy as np
```

```python
open_streets = gpd.read_file("../data/Open_Streets_Locations/geo_export_4a00acf9-46e6-4357-a423-8d64d874a1b7.shp")
```

```python
the_df = pd.DataFrame()
to_list = []
from_list = []
day = []
time = []
open_time = []
for i,j,k,p,opn in zip(open_streets.end_time.tolist(), open_streets.day_of_wee.tolist(), open_streets.from_stree.tolist(),open_streets.to_street.tolist(),open_streets.start_time.tolist()):
    to_list.append(k)
    from_list.append(p)
    splitlist = i.split(',')
    splitlist=[x for x in splitlist if x!='NA']
    opnlist = opn.split(',')
    opnlist=[x for x in opnlist if x!='NA']
    if j=="ALL":
        splitday = ['MTWRFSU']
    else:
        splitday = j.split(',')
    splitday = [j for i in splitday for j in i]
    splitday = [i for i in splitday]

    # End time
    if len(splitday)!=len(splitlist):
        if len(splitlist)<len(splitday):
            splitlist = splitlist + splitlist * (len(splitday)-len(splitlist))
        if len(splitlist)>len(splitday):
            if len(splitlist)==7:
                splitday =['M', 'T', 'W', 'R', 'F', 'S', 'U']
            if len(splitlist)>7:
                splitlist = splitlist[:len(splitday)]
                
                
     # Open time
    if len(splitday)!=len(opnlist):
        if len(opnlist)<len(splitday):
            opnlist = opnlist + opnlist * (len(splitday)-len(opnlist))
            
        if len(opnlist)>len(splitday):
            opnlist = ['5:00PM']* 5 + ['12:00PM']* 2

    splitday = [i for i in splitday if i!='' if i!=' ']
    splitlist = [i for i in splitlist if i!='' if i!=' ']
    
    day.append(splitday)
    time.append(splitlist)
    open_time.append(opnlist)


csvdf = pd.DataFrame(list(zip(from_list, to_list, day, time, open_time)), 
                   columns =['to_street2', 'from_stree2', 'day_open_fixed', 'end_time', 'open_time'])

the_df = the_df.append(csvdf, ignore_index=True)
```

```python
m_values = []
t_values = []
w_values = []
r_values = []
f_values = []
s_values = []
u_values = []
for days, time, tt in zip(the_df.day_open_fixed,the_df.end_time, the_df.to_street2):
    if 'M' in days:
        m_values.append(time[days.index('M')])
    elif 'M' not in days:
        m_values.append(np.nan)
        
    if 'T' in days:
        t_values.append(time[days.index('T')])
    elif 'T' not in days:
        t_values.append(np.nan)
        
    if 'W' in days:
        w_values.append(time[days.index('W')])
    elif 'W' not in days:
        w_values.append(np.nan)
        
    if 'R' in days:
        r_values.append(time[days.index('R')])
    elif 'R' not in days:
        r_values.append(np.nan)
        
    if 'F' in days:
        f_values.append(time[days.index('F')])
    elif 'F' not in days:
        f_values.append(np.nan)
        
    if 'S' in days:
        # Dean Street
        if len(time)==0:
            time = ['10:00PM']
        s_values.append(time[days.index('S')])
    elif 'S' not in days:
        s_values.append(np.nan)
        
    if 'U' in days:
        u_values.append(time[days.index('U')])
    elif 'U' not in days:
        u_values.append(np.nan)
```

```python
csvdf2 = pd.DataFrame(list(zip(m_values, t_values, w_values, r_values, f_values, s_values, u_values)), 
                   columns =['M_end', 'T_end', 'W_end', 'R_end', 'F_end', 'S_end', 'U_end'])
```
